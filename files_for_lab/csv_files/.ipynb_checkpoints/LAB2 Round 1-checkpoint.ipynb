```python
import pandas as pd
import numpy as np
```

```python
# opening files
data1=pd.read_csv('file1.csv')
data2=pd.read_csv('file2.csv')
data3=pd.read_csv('file3.csv')
```

```python
# show the dataframe shape
display(data1.shape)
display(data2.shape)
display(data3.shape)
```

```python
# show the dataframe heads
display(data1.head())
display(data2.head())
display(data3.head())
```

```python
# getting rid of the percentage sign
data1['Customer Lifetime Value']=data1['Customer Lifetime Value'].str.replace("%","")
data2['Customer Lifetime Value']=data2['Customer Lifetime Value'].str.replace("%","")
```

```python
# transforming into numeric
data1['Customer Lifetime Value']=pd.to_numeric(data1['Customer Lifetime Value'], errors='coerce')/100
data2['Customer Lifetime Value']=pd.to_numeric(data2['Customer Lifetime Value'], errors='coerce')/100
```

```python
# chechink
data1.head()
```

```python
# making headers lowercase
data1.columns = data1.columns.str.lower()
data2.columns = data2.columns.str.lower()
data3.columns = data2.columns.str.lower()
```

```python
# Uniforming headers same name
data1=data1.rename(columns={'st': 'state'})
data2=data2.rename(columns={'st': 'state'})
data3=data3.rename(columns={'st': 'state'})
```

```python
# Uniforming headers´ order
data2=data2[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
data3=data3[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
```

```python
#checking
display(data1.head())
display(data2.head())
display(data3.head())
```

```python
#ready to concatenate
dataframe=pd.concat([data1, data2, data3], axis=0)
```

```python
dataframe.shape
```

```python
dataframe.head()
```

```python
#checking types
dataframe.dtypes
```

```python
#changing columns to numeric
dataframe['customer lifetime value']=pd.to_numeric(dataframe['customer lifetime value'], errors='coerce')
dataframe['total claim amount']=pd.to_numeric(dataframe['total claim amount'], errors='coerce')
```

```python
dataframe.dtypes
```

```python
#Deleting columns of the dataframe
dataframe=dataframe.drop(['education', 'number of open complaints'], axis=1)
```

```python
#checking
dataframe.shape
```

```python
 #filtering out the data for customers who have an income of 0 or less.
dataframefilter=dataframe[dataframe['income']==0]
```

```python
#dropping out rows with value 0. Copied code from stackoverflow but I found it a very interesting way of doing it
dataframe = dataframe[dataframe['income'] != 0]
```
